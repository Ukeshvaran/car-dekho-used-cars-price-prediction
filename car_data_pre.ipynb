```python
import pandas as pd
```

```python
df=pd.read_csv('C:/Users/ukesh/OneDrive/Desktop/Python/car_dehko/cleaned_data.csv')
```

```python
df.head(10)
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(df['price'])
plt.show()
```

```python
q1=df['price'].quantile(0.25)
q3=df['price'].quantile(0.75)

iqr=q3-q1
```

```python
q1,q3
```

```python
iqr
```

```python
lwr_limit= q1 - 1.5*iqr
hig_limit= q3 + 1.5*iqr
```

```python
lwr_limit,hig_limit
```

```python
new_df=df[(df['price']>lwr_limit)&(df['price']<hig_limit)]
```

```python
new_df
```

```python
new_df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
sns.boxplot(new_df['price'])
plt.show()
```

```python
new_df
```

```python
new_df.shape
```

```python
new_df.columns
```

```python
num_values=new_df.drop('price',axis=1).select_dtypes(include=['int','float']).columns.to_list()
```

```python
num_values
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
new_df[num_values]=scaler.fit_transform(new_df[num_values])
```

```python
import pickle
with open ('scaler.pkl','wb') as scaler_file:
    pickle.dump(scaler,scaler_file)
```

```python
cat_columns=new_df.select_dtypes(include='object').columns.to_list()
```

```python
cat_columns
```

```python
from sklearn.preprocessing import OneHotEncoder
hot_encoder=OneHotEncoder(sparse_output=False,handle_unknown="ignore") # to keep it as dense array & will handel the unknown data with zero
encoded=hot_encoder.fit(new_df[cat_columns])
```

```python
encoded
```

```python
encoded.categories_
```

```python
encoded_column_names=encoded.get_feature_names_out().tolist()
```

```python
encoded_column_names
```

```python
new_df[encoded_column_names]=hot_encoder.transform(new_df[cat_columns])
```

```python
new_df
```

```python
new_df=new_df.drop(columns=cat_columns,axis=1)
```

```python
with open('OneHotEncoder.pkl','wb') as encoder_file:
    pickle.dump(hot_encoder,encoder_file)
```

```python
new_df.to_csv('Final_data.csv')
```
